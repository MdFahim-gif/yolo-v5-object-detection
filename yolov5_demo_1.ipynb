```python
import os
import shutil
import random

# Directory paths
data_dir = r"C:\Users\HP\yolov5\Crop_Disease_Detection"
train_dir = os.path.join(data_dir, "train")
val_dir = os.path.join(data_dir, "valid")
test_dir = os.path.join(data_dir, "test")

# Ensure directories exist
for dir_path in [train_dir, val_dir, test_dir]:
    os.makedirs(dir_path, exist_ok=True)

# Collect image files
image_files = [file for file in os.listdir(data_dir) if file.endswith('.jpg')]
random.shuffle(image_files)

# Define ratios
train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

# Calculate counts
total_count = len(image_files)
train_count = int(total_count * train_ratio)
val_count = int(total_count * val_ratio)
test_count = total_count - train_count - val_count

# Split datasets
train_files = image_files[:train_count]
val_files = image_files[train_count:train_count + val_count]
test_files = image_files[train_count + val_count:]

# Copy files to respective directories
for file in train_files:
    shutil.copy(os.path.join(data_dir, file), os.path.join(train_dir, file))

for file in val_files:
    shutil.copy(os.path.join(data_dir, file), os.path.join(val_dir, file))

for file in test_files:
    shutil.copy(os.path.join(data_dir, file), os.path.join(test_dir, file))




import yaml

# Load YAML configuration from file
with open( r"C:\Users\HP\yolov5\Crop_Disease_Detection\data.yaml", 'r') as f:
    data_config = yaml.safe_load(f)

# Access the paths and other parameters
train_path = data_config['train']
val_path = data_config['val']
test_path = data_config['test']
nc = data_config['nc']
class_names = data_config['names']


print(f"Training images path: {train_path}")
print(f"Validation images path: {val_path}")
print(f"Test images path: {test_path}")
print(f"Number of classes: {nc}") 
print(f"Class names: {class_names}")


!python train.py --img 415 --batch 8 --epochs 5 --data "C:\Users\HP\yolov5\Crop_Disease_Detection\data.yaml" --weights yolov5s.pt --cache
```
